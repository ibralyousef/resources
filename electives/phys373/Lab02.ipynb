```python
import numpy as np
import matplotlib.pyplot as plt
import sympy
```

## Q1) Definition:

```python
def taylor_series(func, x0, n):
    x = sympy.Symbol('x')
    series = func.series(x, x0=x0, n=n).removeO()
    return sympy.lambdify(x, series)
```

## Q1) Example of e^x:

```python
# Define the function to approximate
x = sympy.Symbol('x')
func = sympy.exp(x)
test=0.5
# Approximate the function at x0 = 0
x0 = 0
n = 10
approx_func = taylor_series(func, x0, n)
error=abs(approx_func(test) -  np.exp(test))
# Print the approximation
print('After %i iterations, the error is %.15f and the result is %.15f'%(n,error,approx_func(test)))
```

## Q2) sin(x) and cos(x)

### For Sin(x) around x0=0 and x1=1.1

```python
# Define the function to approximate
x = sympy.Symbol('x')
func = sympy.sin(x)
value=1
# Approximate the function at x0 = 0
x0 = 0
x1=1.1
# Print the approximation
print("For x = %f" %value)
for n in range(1,10,1):
    y = taylor_series(func, x0, n)
    z=taylor_series(func, x1, n)
    error=abs(y(value) -  np.sin(value))
    error1=abs(z(value) -  np.sin(value))
    print("n=%i    x0=%.1f    error=%.15f       n=%i    x1=%.1f    error=%.15f" %(n,x0,error,n,x1,error1))
x = np.linspace(-2*np.pi,2*np.pi , 100)

labels = ['First Order', 'Third Order', 'Fifth Order', 'Seventh Order']

plt.figure(figsize = (10,8))
y = taylor_series(func, 0, 5)
plt.plot(x,y(x), label=str(5)+str('th order'))
y = taylor_series(func, 0, 10)
plt.plot(x,y(x), label=str(10)+str('th order'))
y = taylor_series(func, 0, 20)
plt.plot(x,y(x), label=str(20)+str('th order'))
plt.plot(x,np.sin(x), label=str('Sin(x)'),color='black')

plt.grid()
plt.title('Taylor Series Approximations of Sin(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

### My choice, and its importance:
The inverse sine function, or arcsine, maps a sine value to its corresponding angle. It has numerous applications in physics, including wave mechanics and control systems. In these fields, a fast and accurate calculation of the inverse sine function is essential.

The importance of using Taylor series to approximate the inverse sine function in physics is its speed and accuracy. This is particularly important in real-time systems and numerical methods for solving differential equations, where a direct calculation of the inverse sine may not be feasible.

In wave mechanics, the inverse sine is used to calculate the phase of a wave from its amplitude. The Taylor series approximation provides a quick and accurate method for this calculation, which is crucial for understanding the behavior of the wave.

In control systems, the inverse sine is used to calculate the angle of a mechanism from its angular velocity. The Taylor series approximation provides a fast and accurate method for determining the angle, which is necessary for ensuring the stability and accuracy of the control system.

In conclusion, the Taylor series approximation of the inverse sine function is a valuable tool in physics due to its speed and accuracy. It has important applications in fields such as wave mechanics and control systems.

```python
x = np.linspace(-1,1 , 100)
y = np.zeros(len(x))
z = np.linspace(-1,-0.75 , 100)
labels = ['First Order', 'Third Order', 'Fifth Order', 'Seventh Order']
func=sympy.asin(sympy.Symbol('x'))
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
for n in range(10,51,10):
    y = taylor_series(func, 0, n)
    ax1.plot(x,y(x), label=str(n)+str('th order'))
    ax2.plot(z,y(z), label=str(n)+str('th order'))

ax1.plot(x, np.arcsin(x), 'k', label = 'ArcSin(x)')
ax1.grid()
ax2.plot(z, np.arcsin(z), 'k', label = 'ArcSin(x)')
ax2.grid()
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax1.legend()
ax2.legend()
ax1.title.set_text('Taylor Series Approximations of ArcSin(x)')
ax2.title.set_text('Zoomed Taylor Series Approximations of ArcSin(x)')
plt.show()
```

### Limitations
Taylor expansions have several limitations when used in computational physics. For example, the accuracy of the approximation depends on the number of terms used in the series. If too few terms are used, the approximation will not be accurate enough, while too many terms may cause the computation to be too slow and inefficient. Additionally, Taylor expansions are not always applicable, as it is difficult to approximate certain functions that are not smooth or have discontinuities.

## Q4) Maclurian for Sin(x)

```python
x = np.linspace(-np.pi,np.pi , 1000)
y = np.zeros(len(x))

labels = ['First Order', 'Third Order', 'Fifth Order', 'Seventh Order']

plt.figure(figsize = (10,8))
for n in range(4):
    y = y + (((-1)**n * (x)**(2*n+1)) / np.math.factorial(2*n+1))
    plt.plot(x,y, label=str(n+1)+str('th order'))

plt.plot(x, np.sin(x), 'k', label = 'Sin(x)')
plt.grid()
plt.title('Maclurian Series Approximations of Various Orders')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
