```python
def newton(x0, tol):
    n=0
    print('x_k    f(x_k)    f"'"(x_k)   f(x_k)/f"'"(x_k)')
    while True:
        n+=1
        x1 = x0 - f(x0)/derivative(x0)
        print('%.4f    %.4f    %.4f   %.4f'%(x0, f(x0),derivative(x0),f(x0)/derivative(x0)))
        if abs(x1-x0) < tol:
            return print('After %i iterations, the solution is %.4f'%(n,x1))
        x0 = x1
        
def f(x):
    return np.exp(-x) - x
def derivative(x):
    return -np.exp(-x) - 1
x0 = 1.0
tol = 0.0001
newton(x0, tol)
```
